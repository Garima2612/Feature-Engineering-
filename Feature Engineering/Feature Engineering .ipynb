### c.	Identify significant variables using p-values and Chi-Square values

```python
from scipy import stats
# Empty lists to store results
significant_vars = []
p_values = []
chi_square_values = []

# Perform chi-square test for each variable
for var in cat_var:
    contingency_table = pd.crosstab(df[var], df[var])
    chi2, p, _, _ = stats.chi2_contingency(contingency_table)
    chi_square_values.append(chi2)
    p_values.append(p)
    if p < 0.05:  # Set the significance level
        significant_vars.append(var)

# Print results
for i in range(len(cat_var)):
    print(f"Variable: {cat_var[i]}")
    print(f"Chi-Square Value: {chi_square_values[i]}")
    print(f"P-value: {p_values[i]}")
    print("Significant: Yes" if cat_var[i] in significant_vars else "Significant: No")
    print()
```

### Finding values of outliers(IQR method)

```python
def detect_outliers_iqr(data):
    outlier_list = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    #print("The Val of Q1 and Q2",q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    #print("The lower & Upper Bound",lwr_bound, upr_bound)
    
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outlier_list.append(i)
    return outlier_list
for i in ['YearBuilt','1stFlrSF','GarageArea','GrLivArea','GarageCars'
          ,'TotRmsAbvGrd','OverallQual','TotalBsmtSF']:
    outliers = detect_outliers_iqr(final_df[i])
    print("Outliers in",i,"attribute :", outliers)
```

### Handling outliers

```python
### Handling outliers using Quantile based flooring and capping method.
###the outlier is capped at a certain value above the 90th percentile value or floored at a factor below the 10th percentile value
###Computing 10th, 90th percentiles and replacing the outliers


def handle_outliers(data):

    tenth_percentile = np.percentile(data, 10)
    ninetieth_percentile = np.percentile(data, 90)

    b = np.where(data<tenth_percentile, tenth_percentile, data)

    b1 = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
    return b1
    
    
for i in ['YearBuilt','1stFlrSF','GarageArea','GrLivArea','GarageCars'
          ,'TotRmsAbvGrd','OverallQual','TotalBsmtSF']:
    final_df[i]=handle_outliers(final_df[i])
```
